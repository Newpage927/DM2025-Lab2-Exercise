### 1.1 Preprocessing Steps

To prepare the raw text data for analysis, I implemented a custom cleaning function `clean_text_improved` that focuses on retaining emotional signals while removing noise:

* **Noise Removal:** Removed URLs, user mentions (`@username`), and HTML tags using regular expressions.
* **Emotional Punctuation Retention:** Unlike standard cleaning which removes all punctuation, I specifically **retained exclamation marks (`!`) and question marks (`?`)**.
    * *Reasoning:* These symbols are strong indicators of intensity and specific emotions (e.g., "!!" for surprise/anger, "?" for confusion/fear).
* **Normalization:** Converted all text to lowercase and stripped extra whitespace.
* **Character Filtering:** Removed non-alphabetic characters (except for the retained punctuation) to reduce vocabulary size.

### 1.2 Feature Engineering Steps

I used **TF-IDF (Term Frequency-Inverse Document Frequency)** to convert text data into numerical vectors, with specific parameters tuned for this dataset:

* **N-gram Range (1, 3):** I included Unigrams, Bigrams, and Trigrams.
    * *Reasoning:* Single words often lack context (e.g., "happy"). Phrases like "not happy" or "really angry" (captured by bi/trigrams) describe the sentiment much more accurately.
* **Max Features (30,000):** Expanded the vocabulary to the top 30,000 features.
    * *Reasoning:* Since I used LinearSVC, which handles high-dimensional sparse data well, increasing the feature space allowed the model to capture more subtle language patterns without suffering from immediate overfitting.
* **Sublinear TF:** Applied `sublinear_tf=True` (logarithmic scaling).
    * *Reasoning:* This reduces the impact of extremely frequent words (like "the", "is") that don't contribute much to sentiment classification.

### 1.3 Explanation of Your Model

For the final submission, I chose the **LinearSVC (Linear Support Vector Classification)** model.

* **Configuration:** I used `C=0.5` to provide stronger regularization compared to the default `C=1.0`, helping to prevent overfitting on the training data.
* **Why LinearSVC?**
    * It is known to scale better to large numbers of samples and features compared to standard SVC.
    * In text classification tasks where the feature space (TF-IDF) is high-dimensional and sparse, LinearSVC often finds the optimal hyperplane more effectively than Logistic Regression.
    * It focuses on the samples that are hardest to classify (support vectors), which is beneficial for distinguishing between similar emotions.

# **Competition Code**

## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section
import pandas as pd
import json
import re

df_id = pd.read_csv('data_identification.csv')
df_emotion = pd.read_csv('emotion.csv')
with open('final_posts.json', 'r', encoding='utf-8') as f:
    data_posts = json.load(f)

posts_list = []
for item in data_posts:
    post_id = item['root']['_source']['post']['post_id']
    text = item['root']['_source']['post']['text']
    posts_list.append({'id': post_id, 'text': text})

df_posts = pd.DataFrame(posts_list)
df_merged = pd.merge(df_id, df_posts, on='id', how='left')
df_merged = pd.merge(df_merged, df_emotion, on='id', how='left')
df_train = df_merged[df_merged['split'] == 'train'].copy()
df_test = df_merged[df_merged['split'] == 'test'].copy()

def clean_text_advanced(text):
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'@\w+', '', text)
    text = re.sub('<.*?>', '', text)
    text = re.sub(r'[^a-zA-Z\s!?]', '', text) 
    text = text.lower()
    text = re.sub(r'\s+', ' ', text).strip()
    return text
df_train['clean_text'] = df_train['text'].apply(clean_text_advanced)
df_test['clean_text'] = df_test['text'].apply(clean_text_advanced)

print("Preprocessing finish")
```

## 2. Feature Engineering Steps

```python
### Add the code related to the feature engineering steps in cells inside this section
from sklearn.feature_extraction.text import TfidfVectorizer

X_train_text = df_train['clean_text']
y_train = df_train['emotion']
X_test_text = df_test['clean_text']

tfidf = TfidfVectorizer(
    max_features=30000,      
    ngram_range=(1, 3),      
    min_df=2,
    stop_words='english',
    sublinear_tf=True 
)

X_train_tfidf = tfidf.fit_transform(X_train_text)
X_test_tfidf = tfidf.transform(X_test_text)

print(f"matrix: {X_train_tfidf.shape}")
```

## 3. Model Implementation Steps

```python
### Add the code related to the model implementation steps in cells inside this section
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_val_score
import pandas as pd

model = LinearSVC(C=0.5, random_state=42, max_iter=3000)
scores = cross_val_score(model, X_train_tfidf, y_train, cv=5, scoring='accuracy', n_jobs=-1)

model.fit(X_train_tfidf, y_train)
print("正在預測測試集...")
y_test_pred = model.predict(X_test_tfidf)

submission_df = pd.DataFrame({
    'id': df_test['id'],
    'emotion': y_test_pred
})

output_filename = 'submission_linearsvc_optimized.csv'
submission_df.to_csv(output_filename, index=False)
```
